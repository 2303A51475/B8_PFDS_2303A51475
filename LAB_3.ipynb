```python
import numpy as np

np.random.seed(0)
A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

add_result = np.add(A, B)
sub_result = np.subtract(A, B)

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nA + B:\n", add_result)
print("\nA - B:\n", sub_result)

manual_add = A[0, 0] + B[0, 0]
print(f"\nManual Check A[0,0] + B[0,0]: {A[0,0]} + {B[0,0]} = {manual_add}")
print(f"From np.add(): {add_result[0,0]}")

print("\nBroadcasting: Not applied here. Both matrices are 3x3 and of equal shape.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([[7, 8, 9],
              [10, 11, 12]])

print("Matrix A (3x2):")
print(A)
print("\nMatrix B (2x3):")
print(B)

C_dot = np.dot(A, B)
print("\nResult of np.dot(A, B):")
print(C_dot)

C_at = A @ B
print("\nResult of A @ B (should be the same as np.dot):")
print(C_at)

image = np.array([[100, 150],
                  [200, 250]])

filter_kernel = np.array([[1, -1],
                          [-1, 1]])

filtered_value = np.sum(image * filter_kernel)
print("\nSimulated 2x2 image:")
print(image)
print("\nFilter kernel:")
print(filter_kernel)
print("\nFiltered result (sum of element-wise multiplication):")
print(filtered_value)

plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(A, cmap='Blues')
plt.title("Matrix A")
plt.colorbar()

plt.subplot(1, 3, 2)
plt.imshow(B, cmap='Greens')
plt.title("Matrix B")
plt.colorbar()

plt.subplot(1, 3, 3)
plt.imshow(C_dot, cmap='Purples')
plt.title("A @ B Result")
plt.colorbar()

plt.tight_layout()
plt.show()
```
